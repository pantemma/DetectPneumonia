Reading the *.csv* file containing the dataset (filenames and corresponding labels).

```
import pandas as pd

dataset = pd.read_csv('/content/drive/MyDrive/labels_train.csv')
```

Splitting the images into 3 directories with the respective class names.

```
import shutil

for i, row in dataset.iterrows():
  fileName = row['file_name']
  classId = row['class_id']
  if classId == 0:
    shutil.move('/content/drive/MyDrive/images/train_images/'+fileName, '/content/drive/MyDrive/images/train_images/0')
  elif classId == 1:
    shutil.move('/content/drive/MyDrive/images/train_images/'+fileName, '/content/drive/MyDrive/images/train_images/1')
  elif classId == 2:
    shutil.move('/content/drive/MyDrive/images/train_images/'+fileName, '/content/drive/MyDrive/images/train_images/2')
  else:
    continue
```

Checking how many images are in each class.

The dataset is unbalanced in favor of the second class.


```
import os

for i in range(3):
  print(i, len(os.listdir(f'/content/drive/MyDrive/images/train_images/{i}')))
```

Previewing the dataset.

Class 0 is normal images, class 1 is bacterial pneumonia and class 2 is viral pneumonia.

Normal image is quite easy to spot, but the difference between bacterial and viral pneumonia is not clear enough.

```
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(12, 6))

for i in range(3):
  imagepath = f'/content/drive/MyDrive/images/train_images/{i}/'
  filename = os.listdir(imagepath)[0]
  img = plt.imread(imagepath+filename)
  fig.add_subplot(1, 3, i+1).set_title(i)
  plt.imshow(img, cmap='gray')
plt.show()
```

Setting the dataset directory, batch size and image size.

```
data_dir = '/content/drive/MyDrive/images/train_images'
batch_size = 64
img_height = 224
img_width = 224
```

Initializing the training set.

```
import tensorflow as tf

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

Initializing the validation set.

```
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

Using buffer prefetching for better performance.

```
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

Augmenting the images to increase their total number as the dataset is not very big.

```
from tensorflow import keras
from tensorflow.keras import layers

data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, 
                                                              img_width,
                                                              3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)
```

Building the model.

```
from tensorflow.keras.models import Sequential

num_classes = 3

model = Sequential([
  data_augmentation,
  layers.experimental.preprocessing.Rescaling(1./255),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(512, activation='relu'),
  layers.Dense(num_classes)
])
```

Compiling the model.

```
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Summarizing the model.

```
model.summary()
```

Setting a checkpont to save the best model during training.

```
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint('model.h5', verbose=1, save_best_only=True)
```

Training the model.

```
epochs=40
history = model.fit(
  train_ds,
  callbacks=[checkpoint],
  validation_data=val_ds,
  epochs=epochs
)
```

Plotting the loss and the accuracy of both validation and training.

```
# plot the loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('Loss')

# plot the accuracy
plt.plot(history.history['accuracy'], label='train accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.legend()
plt.show()
plt.savefig('Accuracy')
```

Loading the best model.

```
from keras.models import load_model
reloaded_model = load_model('/content/model.h5')
```

Predicting images never seen before by the model.

```
from keras.preprocessing import image
import numpy as np
res = []
res.append(['file_name', 'class_id'])

for filename in os.listdir('/content/drive/MyDrive/images/test_images'):
  # predict images
  path = '/content/drive/MyDrive/images/test_images/' + filename
  img = image.load_img(path, target_size=(224,224))
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis =0)

  images = np.vstack([x])
  classes = np.argmax(reloaded_model.predict(images))
  res.append([filename, classes])
```

Writing the prediction results in a *.csv* file.

The accuracy was 80%. Not that bad considering a 3-class problem with 2 very similar classes.


```
import csv

with open('results.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerows(res)
```
